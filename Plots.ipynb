## CNN Test Plot

```python
CNN_SVM = 74.89 
CNN_DT = 66.81 
CNN_KNN = 72.71 
CNN_NB = 33.52 
```

```python
Methods = ['CNN_acc','CNN-SVM','CNN-DT','CNN-KNN','CNN-NB']
accurasies = [CNN_acc,CNN_SVM,CNN_DT,CNN_KNN,CNN_NB]
```

```python
fig = plt.figure(figsize = (10, 5))
 
plt.bar(Methods, accurasies, color ='maroon',width = 0.4)

plt.ylim(0, 100)
 
plt.xlabel("Methods")
plt.ylabel("Accuracy")
plt.title("CNN Accuracies")
plt.show()
```
